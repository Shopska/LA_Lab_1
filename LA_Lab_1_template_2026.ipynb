### What is a transformation matrix?

A transformation matrix is a matrix that represents a linear transformation. When we apply the transformation to a point with coordinates $(x, y)$, we multiply the matrix by the column vector of the coordinates and obtain the new coordinates $(x', y')$. So the matrix encodes how each point in the plane is moved. For example, a scaling matrix multiplies the coordinates by factors along each axis; a rotation matrix rotates points around the origin by a given angle. In this lab we work with 2D transformations, so the matrices are $2 \times 2$ for linear maps (scale, rotation, shear), or we use $3 \times 3$ matrices in homogeneous coordinates when we need translation.

### 1.2 Scaling matrix

Scaling is a tranformation that enlarges or diminishes objects by a factor. For scaling in 2D space a factor can be represented as a vector $(x_s, y_s)$ where $x_s$ and $y_s$ are scaling factors for each respective axis (those factors don't have to be equal). On the picture below, V' represents vector V scaled by a factor (0.5, 1) and V'' - by (0.5, 0.5).
Image 2.png

In this part, you should derive a general form for a scaling matrix and apply scaling by a factor of (1, 0.5) to the vector. First, explain how you derive the scaling matrix:

In two dimensions, scaling means stretching or shrinking a vector along the coordinate axes. 
The direction does not change  we simply multiply each coordinate by a scaling factor.

If the scaling factors are $s_x$ for the x-axis and $s_y$ for the y-axis, then a vector $(x,y)$ transforms into

$$
(x', y') = (s_x x,\; s_y y).
$$

In matrix form, this transformation can be written as

$$
\begin{bmatrix}
x' \\
y'
\end{bmatrix}
=
\begin{bmatrix}
s_x & 0 \\
0 & s_y
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}.
$$

Therefore, the scaling matrix is a diagonal matrix where the diagonal entries are the scaling factors for each axis.


````python
import numpy as np

def get_scaling_matrix(scale_factor_x, scale_factor_y):
    return np.array([
        [scale_factor_x, 0],
        [0, scale_factor_y]
    ])

M2 = get_scaling_matrix(1, 0.5)

print("Scaling matrix is:")
print(M2)
````

Second, apply scaling by (1, 0.5) to the vector.

````python
# Apply scaling to the vector AB1 to achieve vector AB2

A = np.array([0, 0])
B1 = np.array([2, 2])

M = np.array([
    [1, 0],
    [0, 0.5]
])

B2 = M @ B1


ax = plt.axes()

ax.arrow(A[0], A[1], B1[0], B1[1], head_width=0.2, head_length=0.2)
ax.arrow(A[0], A[1], B2[0], B2[1], head_width=0.2, head_length=0.2, color='green')
plt.ylim(-3,3)
plt.xlim(-3,3)

ax.set_axisbelow(True)
plt.grid(color='gray', linestyle='dashed')
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
````

### 1.4 Translation matrix

Translation is a transformation that moves every point of a figure by a fixed distance.


Image 3.png

Here, you should build a translation matrix and apply translation by $(20, 25)$ to the vector. Start from explaining how to construct a translation matrix:

A translation moves every point of a figure by a fixed vector.

If we translate a point $(x, y)$ by $(t_x, t_y)$, then the new coordinates become

$$
x' = x + t_x, \qquad y' = y + t_y.
$$

Unlike scaling or shear, translation cannot be represented by a $2 \times 2$ matrix 
in standard coordinates because it is not a linear transformation (it does not keep the origin fixed).

To represent translation in matrix form, we use homogeneous coordinates:

$$
\begin{bmatrix}
x' \\
y' \\
1
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & t_x \\
0 & 1 & t_y \\
0 & 0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y \\
1
\end{bmatrix}.
$$

So the translation matrix has the translation values in the last column.


Then, implement a function that produces such a matrix:

````python
import numpy as np

def get_translation_matrix(tx, ty):
    return np.array([
        [1.0, 0.0, tx],
        [0.0, 1.0, ty]
    ], dtype=np.float32)

M4 = get_translation_matrix(20, 25)

print("Translation matrix is")
print(M4)
````

### 1.5 Rotation matrix

Rotation is a linear transformation of some space that preserves at least one point. An example of rotation by angle $\psi$ on xy-plane is depicted on the image below:   
Image 1.png

Derive the transformation matrix that corresponds to rotation by a generic angle $\phi$?

To rotate a point $(x, y)$ counterclockwise by an angle $\phi$ around the origin, we use trigonometry.

Let's consider a point at coordinates $(x, y)$ and distance $r$ from the origin, making angle $\theta$ with the x-axis:

$$
x = r \cos\theta, \qquad y = r \sin\theta.
$$

After rotation by angle $\phi$, the new angle becomes $\theta + \phi$, so:

$$
\begin{align*}
x' &= r \cos(\theta + \phi) = r(\cos\theta \cos\phi - \sin\theta \sin\phi) \\
&= x \cos\phi - y \sin\phi, \\
y' &= r \sin(\theta + \phi) = r(\sin\theta \cos\phi + \cos\theta \sin\phi) \\
&= x \sin\phi + y \cos\phi.
\end{align*}
$$

In matrix form:

$$
\begin{bmatrix}
x' \\
y'
\end{bmatrix}
=
\begin{bmatrix}
\cos\phi & -\sin\phi \\
\sin\phi & \cos\phi
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}.
$$

Therefore, the **rotation matrix** for angle $\phi$ (counterclockwise) is:

$$
R(\phi) =
\begin{bmatrix}
\cos\phi & -\sin\phi \\
\sin\phi & \cos\phi
\end{bmatrix}.
$$

Your next task is to implement a function that calculates the rotation matrix and apply it to a vector to rotate it by $45^\circ$.
First, construct the rotation matrix:

````python
# Create rotation matrix

def get_rotation_matrix(phi_degrees):
  ##### YOUR CODE STARTS #####
  phi_radians = np.deg2rad(phi_degrees)
  transformation_matrix = np.array([
      [np.cos(phi_radians), -np.sin(phi_radians)],
      [np.sin(phi_radians),  np.cos(phi_radians)]
  ])
  ##### YOUR CODE ENDS #####
  return transformation_matrix

phi = 45
M = get_rotation_matrix(phi)
print("Rotation matrix is")
print(M)
print(f"Rotation angle: {phi}")
````

Vertical mirror symmetry maps every point $(x, y)$ to $(-x, y)$, reflecting the figure across the $y$-axis.

This means the $x$-coordinate is negated while the $y$-coordinate remains unchanged:

$$
x' = -x, \qquad y' = y.
$$

In matrix form:

$$
\begin{bmatrix}
x' \\
y'
\end{bmatrix}
=
\begin{bmatrix}
-1 & 0 \\
0 & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}.
$$

Therefore, the vertical mirror (reflection over the $y$-axis) matrix is:

$$
M_{\text{mirror}} =
\begin{bmatrix}
-1 & 0 \\
0 & 1
\end{bmatrix}.
$$

### 2.1 Perspective shift (homography) matrix

You will need to use `img2.jpg` for this task. Load and resize it using the same commands you used for preparing img1.

The main goal of perspective transformation (or perspective rectification) is to synthesize a perspective-free view of an object given a photo taken at an angle.   

We recommend reading more here: https://docs.opencv.org/4.x/d9/dab/tutorial_homography.html.

Untitled Diagram.jpg

As you see on the picture above, the initial picture has a standard basis of red vectors. We want to transform all of the vectors in the picture so that the blue vectors will become the new basis.

To find this transformation matrix, you need to choose 4 points on the input image and corresponding points on the output image ($3$ of these points should not be collinear). Simply saying, you can choose these points using, for example, Paint.

Our goal is to find a transformation that maps each of our initial 4 points $(x_i,y_i)$ to 4 points with new coordinates $(u_i, v_i)$.

In this case, the transformation matrix is a $3 \times 3$ matrix that satisfies the following equation, where ($x_i$, $y_i$) is the source vector and ($u_i$, $v_i$) is  the destination vector:

$$
\begin{bmatrix}
u_i \\
v_i \\
1
\end{bmatrix} =
\begin{bmatrix}
c_{00} & c_{01} & c_{02} \\
c_{10} & c_{11} & c_{12} \\
c_{20} & c_{21} & c_{22}
\end{bmatrix}
\begin{bmatrix}
x_i \\
y_i \\
1
\end{bmatrix}
$$



Slightly rewriting the equation we can move to:

$$ u_i = \dfrac{c_{00}*x_i + c_{01}*y_i + c_{02}}{ c_{20}*x_i + c_{21}*y_i + c_{22}} $$

$$ v_i = \dfrac{c_{10}*x_i + c_{11}*y_i + c_{12}}{c_{20}*x_i + c_{21}*y_i + c_{22}} $$

where $\textcolor{red}{(u_i, v_i)}$ is the coordinates of the destination points and $\textcolor{blue}{(x_i, y_i)}$ is coodinates of the sourse points. In our case sourse points are that on the paper(blue vector) and destination is just corners of the photo (red vectors).


Find coefficients $c_{ij}$ via solving the linear system containing source and destination coordinates and nine $c$ coefficients with numpy methods and apply it to an image so that the sign is clearly visible.

(Hint: $c_{22}$ can be taken as any number; for example, $c_{22} = 1$; the reason is that the matrix we are looking for has $8$ degrees of freedom as it is estimated up to a constant)

First, start from explaining how to move from the equations for $u_i$ and $v_i$ to a system of linear equations of the form $A\mathbf{x} = \mathbf{b}$.

We set $c_{22} = 1$ (so the matrix is determined up to scale; 8 degrees of freedom). Multiply both sides of $u_i = \frac{c_{00}x_i + c_{01}y_i + c_{02}}{c_{20}x_i + c_{21}y_i + 1}$ by the denominator: $u_i(c_{20}x_i + c_{21}y_i + 1) = c_{00}x_i + c_{01}y_i + c_{02}$. Rearranging: $c_{00}x_i + c_{01}y_i + c_{02} - u_i c_{20} x_i - u_i c_{21} y_i = u_i$. So we get one linear equation in the 8 unknowns $(c_{00}, c_{01}, c_{02}, c_{10}, c_{11}, c_{12}, c_{20}, c_{21})$ per point for $u$, and similarly for $v$: $c_{10}x_i + c_{11}y_i + c_{12} - v_i c_{20} x_i - v_i c_{21} y_i = v_i$. With 4 point pairs we get 8 equations. Writing $\mathbf{x} = [c_{00}, c_{01}, c_{02}, c_{10}, c_{11}, c_{12}, c_{20}, c_{21}]^T$, the system is $A\mathbf{x} = \mathbf{b}$ where each row of $A$ comes from one equation (e.g. for $u_i$ the row is $[x_i, y_i, 1, 0, 0, 0, -u_i x_i, -u_i y_i]$ with $b_i = u_i$).

Do the outputs coincide? Is it possible to obtain same/different result changing the order of transformations? Which conditions should be met? Derive it theoretically.

The two outputs do not coincide. Matrix multiplication is not commutative, so the order of multiplication affects the result.

Transformations are represented as matrices $S$, $R$, and $T$:

$$
S = \begin{bmatrix} 
1.5 & 0 & 0 \\ 
0 & 2 & 0 \\ 
0 & 0 & 1 
\end{bmatrix}
$$

$$
R = \begin{bmatrix} 
\cos 18^\circ & -\sin 18^\circ & 0 \\ 
\sin 18^\circ & \cos 18^\circ & 0 \\ 
0 & 0 & 1 
\end{bmatrix}
$$

$$
T = \begin{bmatrix} 
1 & 0 & 10 \\ 
0 & 1 & 15 \\ 
0 & 0 & 1 
\end{bmatrix}
$$

If $S$, $R$ and $T$ are matrices of order $m \times m$, then in general $TRS \neq RST$. So the two combined matrices $M_1 = TRS$ and $M_2 = RST$ are different and the images do not coincide.

It is still possible to get the same result for different orders in special cases: if we apply only two scalings or only two translations, their order does not matter.

Each transformation can be written as a homogeneous matrix. When we apply $T_1$ to a point $P = (x, y, 1)$ and then $T_2$, we get $P'' = T_2 \cdot (T_1 \cdot P) = (T_2 \cdot T_1) \cdot P$. If we apply $T_2$ first and then $T_1$, we get $P'' = (T_1 \cdot T_2) \cdot P$. So the combined transformation depends on the order: we get $T_2 T_1$ in one case and $T_1 T_2$ in the other. For the two results to be the same we would need $T_1 \cdot T_2 = T_2 \cdot T_1$, which in general does not hold.

Proof that scalings commute:

A scaling matrix has the form $S(s_x, s_y) = \begin{bmatrix} s_x & 0 & 0 \\ 0 & s_y & 0 \\ 0 & 0 & 1 \end{bmatrix}$. For two scalings $S_1$ and $S_2$ with factors $(s_1, t_1)$ and $(s_2, t_2)$ we have

$$S_1 \cdot S_2 = \begin{bmatrix} s_1 & 0 & 0 \\ 0 & t_1 & 0 \\ 0 & 0 & 1 \end{bmatrix} \cdot \begin{bmatrix} s_2 & 0 & 0 \\ 0 & t_2 & 0 \\ 0 & 0 & 1 \end{bmatrix} = \begin{bmatrix} s_1 s_2 & 0 & 0 \\ 0 & t_1 t_2 & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

and

$$S_2 \cdot S_1 = \begin{bmatrix} s_2 & 0 & 0 \\ 0 & t_2 & 0 \\ 0 & 0 & 1 \end{bmatrix} \cdot \begin{bmatrix} s_1 & 0 & 0 \\ 0 & t_1 & 0 \\ 0 & 0 & 1 \end{bmatrix} = \begin{bmatrix} s_2 s_1 & 0 & 0 \\ 0 & t_2 t_1 & 0 \\ 0 & 0 & 1 \end{bmatrix}.$$

Since multiplication of numbers is commutative, $s_1 s_2 = s_2 s_1$ and $t_1 t_2 = t_2 t_1$, so $S_1 \cdot S_2 = S_2 \cdot S_1$. Thus scaling transformations commute.

Proof that translations commute:

A translation matrix has the form $T(x_0, y_0) = \begin{bmatrix} 1 & 0 & x_0 \\ 0 & 1 & y_0 \\ 0 & 0 & 1 \end{bmatrix}$. For two translations $T_1(x_1, y_1)$ and $T_2(x_2, y_2)$:

$$T_1 \cdot T_2 = \begin{bmatrix} 1 & 0 & x_1 \\ 0 & 1 & y_1 \\ 0 & 0 & 1 \end{bmatrix} \cdot \begin{bmatrix} 1 & 0 & x_2 \\ 0 & 1 & y_2 \\ 0 & 0 & 1 \end{bmatrix} = \begin{bmatrix} 1 & 0 & x_1 + x_2 \\ 0 & 1 & y_1 + y_2 \\ 0 & 0 & 1 \end{bmatrix}.$$

Swapping the order gives $T_2 \cdot T_1$ with last column $(x_2 + x_1, y_2 + y_1, 1)^T$. Since addition is commutative, $x_1 + x_2 = x_2 + x_1$ and $y_1 + y_2 = y_2 + y_1$, so $T_1 \cdot T_2 = T_2 \cdot T_1$. Thus translation transformations commute.

In NumPy, a **grayscale image** is represented as a **2D matrix**, where each element corresponds to the intensity of a pixel $I_{gray} \in \mathbb{R}^{H \times W}$,where $H$ and $W$ are the image height and width, respectively.

By working with grayscale images, we reduce the problem to single-channel interpolation, making it easier to comprehend at the beginning

One of the simplest image inpainting methods for small missing areas is **neighbor averaging**, where a missing pixel is restored by computing the average value of its known neighboring pixels.

For a missing pixel $I(x, y)$, its value can be estimated as:

$$
I(x, y) = \frac{1}{|(x, y)|} \sum_{(i,j) \in (x, y)} I(i, j)
$$

where:  
- $(x, y)$ is the set of known neighboring pixels around $(x, y)$  
- $|(x, y)|$ is the number of valid neighbors  
- $I(i, j)$ is the intensity value of a neighboring pixel

---

The set of known neighboring pixels $(x, y)$ depends on:
* whether we consider 4-connectivity (direct neighbors only):
$$
_{4}(x, y) = \{ I(x-1, y), I(x+1, y), I(x, y-1), I(x, y+1) \}
$$

* or 8-connectivity (including diagonal neighbors):
$$
_{8}(x, y) = \{ I(x-1, y-1), I(x-1, y), I(x-1, y+1), I(x, y-1), I(x, y+1), I(x+1, y-1), I(x+1, y), I(x+1, y+1) \}
$$

---

In a **33 grid**, the missing pixel at $(1,1)$ is restored using its **top, bottom, left, and right** neighbors:
$$
I(1,1) = \frac{I(0,1) + I(1,0) + I(1,2) + I(2,1)}{4}
$$

With **all 8 neighbors (including those on the angles)**, the missing pixel is estimated as:

$$
I(1,1) = \frac{I(0,0) + I(0,1) + I(0,2) + I(1,0) + I(1,2) + I(2,0) + I(2,1) + I(2,2)}{8}
$$


````python
# 3x3 image with a missing center pixel
original_image = np.array([
    [50, 120, 40],
    [90, None, 160],
    [30, 130, 20]
], dtype=object)

# Restore using only direct (4) neighbors
direct_neighbors = [original_image[i, j] for i, j in [(0,1), (1,0), (1,2), (2,1)] if original_image[i, j] is not None]
restored_image_4 = original_image.copy()
restored_image_4[1, 1] = int(np.mean(direct_neighbors))

# Restore using all (8) neighbors
all_neighbors = [original_image[i, j] for i in range(3) for j in range(3) if original_image[i, j] is not None]
restored_image_8 = original_image.copy()
restored_image_8[1, 1] = int(np.mean(all_neighbors))

def plot_image(image, title, ax):
    ax.set_xlim(0, 3)
    ax.set_ylim(0, 3)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(title, fontsize=10)

    for i in range(3):
        for j in range(3):
            value = image[i, j]
            color = (0, 0, value / 255) if value is not None else (1, 1, 1)  # white for missing pixel
            rect = patches.Rectangle((j, 2 - i), 1, 1, facecolor=color, edgecolor='black', linewidth=2)
            ax.add_patch(rect)

fig, axs = plt.subplots(1, 3, figsize=(9, 3))
plot_image(original_image, "Before (missing pixel)", axs[0])
plot_image(restored_image_4, "After (4 neighbors)", axs[1])
plot_image(restored_image_8, "After (8 neighbors)", axs[2])
plt.show()
````

#### 3.2.1

````python
# Construct the linear system Ax = B for missing pixels
def construct_linear_system(pixel_matrix, damaged_coords):
    """
    Explanations:
    - This function constructs a linear system where each missing pixel is represented by an equation.
    - Each missing pixel is expressed as the average of its valid neighboring pixels.
    - The mapping 'var_index' assigns a unique index to each missing pixel, allowing it to be referenced in matrix form.

    Example:
    var_index = {
        (2, 3): 0, # first missing pixel -> X_0
        (3, 4): 1, # second missing pixel -> X_1
        (5, 2): 2  # third missing pixel -> X_2
    }

    Returns:
    - A (list of lists): Coefficient matrix representing the system of equations.
    - b (list): Right-hand side vector containing known pixel values.
    """
    height, width = len(pixel_matrix), len(pixel_matrix[0])
    var_index = {coord: i for i, coord in enumerate(damaged_coords)}
    n = len(damaged_coords)
    A = [[0] * n for _ in range(n)]
    b = [0] * n

    ### YOUR CODE SOLUTION START
    deltas = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    for i, (y, x) in enumerate(damaged_coords):
        known_sum = 0.0
        n_neighbors = 0
        for dy, dx in deltas:
            ny, nx = y + dy, x + dx
            if 0 <= ny < height and 0 <= nx < width:
                n_neighbors += 1
                if (ny, nx) in var_index:
                    j = var_index[(ny, nx)]
                    A[i][j] -= 1.0
                else:
                    known_sum += float(pixel_matrix[ny][nx])
        if n_neighbors == 0:
            continue
        A[i][i] = 1.0
        for j in range(n):
            A[i][j] /= n_neighbors
        A[i][i] = 1.0
        b[i] = known_sum / n_neighbors
    ### YOUR CODE SOLUTION END

    return A, b
````

#### 3.2.2

````python
# Solve the linear system using Gaussian elimination or NumPy
def solve_linear_system(A, b, method='numpy'):
    """
    Solves the linear system Ax = B.

    Parameters:
    - A (list of lists): Coefficient matrix.
    - b (list): Right-hand side vector.
    - method (str): 'numpy' to use NumPy solver, 'gaussian' to use manual Gaussian elimination.

    Returns:
    - x (list): Solution vector containing the restored pixel values.
    """
    if method == 'numpy':
        return np.linalg.solve(np.array(A, dtype=float), np.array(b, dtype=float))
    elif method in ('gaussian', 'manual'):
        n = len(A)
        A = [row[:] for row in A]
        b = list(b)
        for i in range(n):
            A[i] = [float(a) for a in A[i]]
            b[i] = float(b[i])

        ### YOUR CODE SOLUTION START
        for col in range(n):
            pivot_row = col
            for r in range(col + 1, n):
                if abs(A[r][col]) > abs(A[pivot_row][col]):
                    pivot_row = r
            A[col], A[pivot_row] = A[pivot_row], A[col]
            b[col], b[pivot_row] = b[pivot_row], b[col]
            pivot = A[col][col]
            if abs(pivot) < 1e-12:
                raise ValueError("Singular matrix in Gaussian elimination")
            for r in range(col + 1, n):
                factor = A[r][col] / pivot
                for c in range(col, n):
                    A[r][c] -= factor * A[col][c]
                b[r] -= factor * b[col]
        x = [0.0] * n
        for i in range(n - 1, -1, -1):
            x[i] = b[i]
            for j in range(i + 1, n):
                x[i] -= A[i][j] * x[j]
            x[i] /= A[i][i]
        ### YOUR CODE SOLUTION END

        return x
    else:
        raise ValueError(f"Unknown method: {method}")
````

````python
def complete_solution(img_gray, damaged_matrix, damaged_coords, method='numpy'):
  try:
      solution = solve_linear_system(A, b, method)
  except ValueError as e:
      print("Error solving the system:", e)
      solution = None

  # Apply the reconstructed pixel values
  if solution is not None:
      restored_matrix = [row[:] for row in damaged_matrix]
      for i, (y, x) in enumerate(damaged_coords):
          restored_matrix[y][x] = int(solution[i])

      # Create visualization
      restored_img = set_pixel_matrix(img_gray.copy(), restored_matrix)
      damaged_img = set_pixel_matrix(img_gray.copy(), damaged_matrix)
      highlighted_matrix = [row[:] for row in damaged_matrix]
      for (y, x) in damaged_coords:
          highlighted_matrix[y][x] = 255
      highlighted_img = set_pixel_matrix(img_gray.copy(), highlighted_matrix)

      # Display results
      fig, axs = plt.subplots(1, 3, figsize=(15, 5))
      axs[0].imshow(img_gray, cmap='gray')
      axs[0].set_title("Original Image")
      axs[1].imshow(highlighted_img, cmap='gray')
      axs[1].set_title("Damaged Image")
      axs[2].imshow(restored_img, cmap='gray')
      axs[2].set_title("Reconstructed Image (" + method + ")")

      for ax in axs:
          ax.axis("off")

      plt.show()
````

1. Other methods to solve linear systems: Besides Gaussian elimination and NumPy's solver (which typically uses LAPACK, e.g. LU or similar), one can use LU decomposition (factor $A=LU$ then solve two triangular systems), iterative methods (Jacobi, GaussSeidel, conjugate gradient) for large sparse systems, Cholesky when $A$ is symmetric positive definite, and matrix inversion $A^{-1}\mathbf{b}$ (usually avoided for numerical reasons).

2. Damaged pixels at the edges: At the image boundary, some 4-neighbors are out of bounds, so $|\mathcal{N}(x,y)|$ is smaller (2 or 3). The same formulas apply: we still have $I(x,y) = \frac{1}{|\mathcal{N}|}\sum I(i,j)$, but the row of $A$ has fewer nonzero off-diagonal terms and $b$ uses only the in-bounds known neighbors. The system remains well-defined as long as each missing pixel has at least one known neighbor.

3. Diagonal neighbors (8-connectivity): Each pixel equation would use up to 8 neighbors, so $|\mathcal{N}|$ can be up to 8. The matrix $A$ becomes denser (more off-diagonal entries per row), and $b$ includes contributions from diagonal known pixels. The solution can differ because the averaging is over a larger neighborhood and thus smoother.

4. Small vs large damage: For small damage, unknowns are surrounded by known pixels, so the local average is a good estimate. For large areas, many unknowns depend mainly on each other, so the solution is more like a smooth interpolation with little anchor from the image, and texture or sharp edges are lost.

5. Improving for complex damage: One can use patch-based or exemplar-based inpainting (copy patches from known regions), variational/PDE methods (e.g. harmonic or biharmonic extension), deep learning (e.g. GANs or diffusion models trained to fill holes), or edge-aware diffusion so that filling respects boundaries and textures.

### How its shape and values affect the resulting image? What is a shape of a typical transformation matrix that is applied to images; which factors affect its shape?

For 2D images we work with 2D points $(x, y)$. A typical transformation matrix is $2 \times 2$ when we only do linear stuff (scaling, rotation, shear): it multiplies the column vector $[x, y]^T$ and gives new coordinates. If we also want translation, we use homogeneous coordinates and then the matrix is $3 \times 3$: the extra row/column allow the constant shift. So the shape depends on whether we use ordinary or homogeneous coordinates. The values in the matrix set what happens: diagonal entries give scaling, off-diagonal can give rotation/shear, and in the $3 \times 3$ case the last column (except the bottom right 1) gives translation. So the same shape of matrix (e.g. $3 \times 3$) can give very different results (e.g. only translation vs only rotation) depending on the numbers we put in.

### What can be a result of composition of transformations?

Composition of transformations is when we apply one transformation after another (e.g. scaling then rotation). The result is again a transformation: it corresponds to one matrix  the product of the matrices of those transformations. So the composition gives another linear/affine transformation that we can write as a single matrix. The order usually matters: scale then translate is not the same as translate then scale.

### What is a perspective shift transformation in terms of geometry?

A perspective shift in geometric terms is when the flat image is transformed as if we are looking at it from a different angle. Parallel lines in the image after this transformation are no longer necessarily parallel (e.g. a rectangle can look like a trapezoid). So it is not just translation or rotation, but a change of the "viewing angle" onto the plane.
